```python
import pandas as pd
import numpy as np

synthetic=pd.read_csv("synthetic_personal_finance_dataset.csv")
synthetic.head()
```

```python
new_columns=['age','gender','education_level','job_title','monthly_income_usd','monthly_expense_usd','savings_usd''debt_to_income_ration','credit_score','savings_income_ration']
synthetic=synthetic.drop([c for c in synthetic.columns if c not in new_columns],axis=1)
synthetic.head()
```

```python
import os
data=synthetic
df=pd.DataFrame(data)
csv_filename="cleaned_synthetic.csv"
df.to_csv(csv_filename, index=False)
print(f"Success! '{csv_filename}' has been created in: {os.getcwd()}")
```
