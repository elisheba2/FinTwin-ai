```python
import pandas as pd
import numpy as np

personalexpense=pd.read_csv("personal_expense_dataset.csv")
personalexpense.head()
```

```python
personalexpense.isnull().sum()
```

```python
new_columns=['Category','Amount','Month']
personalexpense=personalexpense.drop([c for c in personalexpense.columns if c not in new_columns],axis=1)
personalexpense.head()
```

```python
import os
data=personalexpense
df=pd.DataFrame(data)
csv_filename="cleaned_personalexpense.csv"
df.to_csv(csv_filename, index=False)
print(f"Success! '{csv_filename}' has been created in: {os.getcwd()}")
```
