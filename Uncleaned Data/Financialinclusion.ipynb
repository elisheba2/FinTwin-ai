```python
import pandas as pd
import numpy as np

financialinclusion=pd.read_csv("Financial_inclusion_dataset.csv")
financialinclusion.head()
```

```python
financialinclusion.isnull().sum()
```

```python
financialinclusion.head()
```

```python
new_columns=['bank_account','location_type','age_of_respondent','education_level','job_type']
financialinclusion=financialinclusion.drop([c for c in financialinclusion.columns if c not in new_columns],axis=1)
financialinclusion.head()
```

```python
import os
data=financialinclusion
df=pd.DataFrame(data)
csv_filename="cleaned_financialinclusion.csv"
df.to_csv(csv_filename, index=False)
print(f"Success! '{csv_filename}' has been created in: {os.getcwd()}")
```
