```python
import pandas as pd
import numpy as np

personalfinance=pd.read_csv("personal_finance_tracker_dataset.csv")
personalfinance.head()
```

```python
personalfinance.isnull().sum()
```

```python
personalfinance.head()
```

```python
new_columns = ['monthly_income', 'monthly_expense_total', 'savings_rate','budget_goal','credit_score','debt_to_income_ratio','loan_payment','discretionary_spending','essential_spending','rent_or_mortgage', 'financial_advice_score', 'financial_stress_level','actual_savings']
personalfinance=personalfinance.drop([c for c in personalfinance.columns if c not in new_columns],axis=1)
personalfinance.head()
```

```python
import os
data=personalfinance
df=pd.DataFrame(data)
csv_filename="cleaned_personalfinance.csv"
df.to_csv(csv_filename, index=False)
print(f"Success! '{csv_filename}' has been created in: {os.getcwd()}")
```
