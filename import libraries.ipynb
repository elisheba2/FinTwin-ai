```python
df.columns = (
    df.columns
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

df.columns
```

```python
duplicate_columns = df.columns[
    df.columns.duplicated()
]

duplicate_columns
```

```python
numeric_columns = df.select_dtypes(include='number').columns

numeric_columns

numeric_columns
```

```python
for column in df.select_dtypes(include='number').columns:
    df[column] = df[column].fillna(df[column].median())
```

```python
financial_columns = [
    "Monthly_Income",
    "Monthly_Expenses",
    "Monthly_Savings",
    "Loan_Amount",
    "Debt",
    "Bank_Balance"
]


for column in financial_columns:
    if column in df.columns:
        print(
            column,
            "negative values:",
            (df[column] < 0).sum()
        )
```

```python
import numpy as np

for column in financial_columns:
    if column in df.columns:
        df[column] = np.where(
            df[column] < 0,
            0,
            df[column]
        )
```

```python
print("Missing values:")
print(df.isnull().sum().sum())


print("\nDuplicates:")
print(df.duplicated().sum())


print("\nShape:")
print(df.shape)
```
