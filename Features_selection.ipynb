```python
from sklearn.ensemble import RandomForestRegressor
```

```python
# specify the features and target variable
target = "Financial_Health_Score"
```

```python
# selecting features and target variable
x = df.drop(target, axis = 1)
y = df[target]
```

```python
x = pd.get_dummies(x, drop_first=True)  # convert categorical variables to dummy variables
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=100, random_state=42)
```

```python
importances = pd.Series(rf.fit(x, y).feature_importances_, index=x.columns).sort_values(ascending=False)
print(importances)
```

```python
strong_predictors = importances[importances > 0.05].index.tolist()
print("Strong predictors:", strong_predictors)
```

```python
print("Top 10 Predictors:")
print("----------------------------")
for index, row in importances.head(10).reset_index().iterrows():
    print(f"{index + 1}. {row['index']}: {row[0]:.4f}") 
    print()
```
